```python
import requests
import pandas as pd
import datetime as dt
from datetime import date, timedelta
import yfinance as yf
```

```python
# Initialize DataFrame for BTC-CAD and ETH-CAD

# Set datetime format for one year
sd = dt.datetime(2022, 3, 14)
ed = dt.datetime(2023, 3, 14)

# Create a DataFrame for BTC-CAD and ETH-CAD
BTC_ETH = yf.Tickers("BTC-CAD ETH-CAD")
BTC_ETH_1d = BTC_ETH.history(period = "1d")
BTC_ETH_1d
```

```python
# Pick BTC-CAD and ETH-CAD close prices
btc_close_price = float(BTC_ETH_1d["Close"]["BTC-CAD"])
eth_close_price = float(BTC_ETH_1d["Close"]["ETH-CAD"])

# Print BTC-CAD and ETH-CAD close prices
print(f"Current AGG closing price: ${btc_close_price:0.2f}")
print(f"Current SPY closing price: ${eth_close_price:0.2f}")
```

```python
# Get 5 years' worth of historical data for BTC-CAD and ETH-CAD
# Reorganize the DataFrame
# Separate ticker data
btc = pd.DataFrame(yf.Ticker("BTC-CAD").history(start = sd, end = ed))
btc.rename(columns = {"Open":"open", "High":"high", "Low":"low", "Close":"close", "Volume":"volume", "Dividends":"dividend", "Stock Splits":"stock splits"}, inplace = True)
eth = pd.DataFrame(yf.Ticker("ETH-CAD").history(start = sd, end = ed))
eth.rename(columns = {"Open":"open", "High":"high", "Low":"low", "Close":"close", "Volume":"volume", "Dividends":"dividend", "Stock Splits":"stock splits"}, inplace = True)

# Concatenate the ticker DataFrames
df_crypto_data = pd.concat([btc, eth], axis=1, keys=["BTC-CAD", "ETH-CAD"])

# Display sample data
df_crypto_data.head()
```
